```python
#execution based on a certain condition

#ZERO BASED INDEXING ALWAYS APPLIES

temperatures = [26.8, 30.5, 33, 34, 27.9, 31.2, 31.9, 29, 30.7, 32]

index = int(input("Enter an index "))

print("The temperature at index", index , "is", temperatures[index])

if temperatures[index] >= 30:
  print("It is hot")
elif 25 <= temperatures[index] < 30:
  print("It's the perfect weather!")
else:
  print("It is cold")


#else if, written as elif
```

###`WHILE`


```python
#perform an action while a certain condition is true

#to read 5 temperature records, a while loop has been used

#while a condition is true, do something

count = 0

five_temperature_records = []

while count < 5:
  temp = float(input("Enter the temperature of record "))
  
  #keep track of this variable count, as soon as it reaches a value 5, while loop stops execution
  print("The value of count is ", count)

  five_temperature_records.append(temp)
  
  #count = count + 1

print(five_temperature_records)
```

###`FUNCTIONS`

```python
#consider performing this task repetitively

celsius_1 = float(input("Enter temperature in celsius "))
fahrenheit_1 = (celsius_1 * (9/5)) + 32
print("Temperature in celsius is", celsius_1, "and temperature in fahrenheit is", fahrenheit_1)

celsius_2 = float(input("Enter temperature in celsius "))
fahrenheit_2 = (celsius_2 * (9/5)) + 32
print("Temperature in celsius is", celsius_2, "and temperature in fahrenheit is", fahrenheit_2)

celsius_3 = float(input("Enter temperature in celsius "))
fahrenheit_3 = (celsius_3 * (9/5)) + 32
print("Temperature in celsius is", celsius_3, "and temperature in fahrenheit is", fahrenheit_3)

# ELIMINATE REDUNDANT CODE
```

image.png

```python
#simple function definition

#take in the temperature in celsius as a PARAMETER

#RETURN is used to return the value after computation

#def refers to defining a function

def celsius_to_fahrenheit( temp_in_celsius ): 

  temp_in_fahrenheit = (temp_in_celsius * (9/5)) + 32 
  
  return temp_in_fahrenheit 

temp_in_celsius = float(input("Enter temperature in celsius "))

temp_in_fahrenheit = celsius_to_fahrenheit(temp_in_celsius)

print("Temperature in celsius is", temp_in_celsius, "and temperature in fahrenheit is", temp_in_fahrenheit)
```

### `CLASSES AND OBJECTS`

Related variables and functions are under the same umbrella


image.png

```python
#associated variables and functions are put together
#use of variables and associated functions become more clear

class weather:
  
  def __init__(self):
    self.temp_in_celsius = 0.0
    self.temp_in_fahrenheit = 0.0

  def set_temperature(self):
    self.temp_in_celsius = float(input("Enter the temperature "))

  def celsius_to_fahrenheit(self):
    self.temp_in_fahrenheit = (self.temp_in_celsius * (9/5)) + 32 
    
  def get_data(self):
    print("The temperature is celsius is", self.temp_in_celsius)
    print("The temperature in fahrenheit is", self.temp_in_fahrenheit)
```

```python
#creating an object of class weather
#this is for TODAY

#create the object
todays_weather = weather()

#call its associated functions

todays_weather.set_temperature()

todays_weather.celsius_to_fahrenheit()

todays_weather.get_data()
```

```python
#for yesterday's weather

yesterdays_weather = weather()

yesterdays_weather.set_temperature()

yesterdays_weather.celsius_to_fahrenheit()

yesterdays_weather.get_data()

#considering a case where we have to 
```
